```python
import seaborn as sns 
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,palette='coolwarm',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,palette='coolwarm',hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',palette='magma')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,row='sex',col='time',palette='magma')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='day',palette='magma')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='day',palette='magma',aspect=0.6,height=8)
```
