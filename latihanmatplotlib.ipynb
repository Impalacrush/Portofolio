```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(1,5,11)
```

```python
x
```

```python
y=x**2
```

```python
y
```

```python
plt.plot(x,y,'g')
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.title('title table')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b--')
plt.subplot(1,2,2)
plt.plot(y,x,'r*--')
```

```python
fig= plt.figure()

axes=fig.add_axes([0.1 , 0.1 , 0.8,0.8])

axes.plot(x,y,'b')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('kikuchi')
```

```python
fig= plt.figure()

axes1=fig.add_axes([0.1 , 0.1 , 0.8,0.8])

axes2=fig.add_axes([0.2 , 0.5 , 0.4,0.3])
axes1.plot(x,y,'r')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('kikuchi')
axes2.plot(x,y,'y')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('naruyoshi')
```

```python
fig=plt.figure(figsize=(8,4),dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(8,4),dpi=100)

axes.plot(x,y,'r')
axes.set_xlabel('hai')
axes.set_ylabel('hello')
axes.set_title('its odb')

fig.savefig('odb.png',dpi=200)
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='data hellopangkat2')
axes.plot(x,x**3,label='data haipangkat3')
axes.legend(loc=1)
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])
axes.plot(x,x+1,color='gray',linewidth=0.25)
axes.plot(x,x+2,color='purple',linewidth=0.50)
axes.plot(x,x+3,color='yellow',linewidth=0.75)
axes.plot(x,x+4,color='brown',linewidth=1)
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])
axes.plot(x,x+1,color='green',linewidth=3,ls='solid')
axes.plot(x,x+2,color='cyan',linewidth=3,ls=':')
axes.plot(x,x+3,color='magenta',linewidth=3,ls='-.')
axes.plot(x,x+4,color='pink',linewidth=3,linestyle='-')
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])
line,=axes.plot(x,x+1,color='red',linewidth=1.50)
line.set_dashes([5,5,15,5])
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])

axes.plot(x,x+1,color='cyan',linewidth=3,ls='-',marker='o')
axes.plot(x,x+2,color='cyan',linewidth=3,ls=':')
axes.plot(x,x+3,color='magenta',linewidth=3,ls='-.')
axes.plot(x,x+4,color='pink',linewidth=3,linestyle='-')
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])

axes.plot(x,x+1,color='cyan',linewidth=3,ls='-',marker='o',markersize='10')
axes.plot(x,x+2,color='cyan',linewidth=3,ls=':',markerfacecolor='red')
axes.plot(x,x+3,color='magenta',linewidth=3,ls='-',marker='o',markerfacecolor='yellow',markeredgewidth='3',markeredgecolor='green')
axes.plot(x,x+4,color='pink',linewidth=3,linestyle='-')
```
