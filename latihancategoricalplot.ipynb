```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='coolwarm')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='coolwarm',hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='coolwarm',hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',palette='Set2',dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',palette='rainbow',dodge=True)
```

```python
sns.violinplot(x='tip',y='day',data=tips,palette='rainbow')
sns.swarmplot(x='tip',y='day',data=tips,palette='Set2',size=2)
```

```python
sns.catplot(x='sex',y='total_bill',data=tips,hue='day',dodge=True,kind='bar')
```
